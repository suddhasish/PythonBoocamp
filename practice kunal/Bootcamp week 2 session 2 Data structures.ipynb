```python
# objective : extract marks, calculate total score, average
marks = 'maths 98 english 77 science 60'
```

```python
marks_list = marks.split()
marks_list
```

```python
only_marks = []
for i in marks_list:
    if i.isdigit():
        only_marks.append(int(i))
only_marks
```

```python
print("total marks : ", sum(only_marks))
print("total marks : ", sum(only_marks)/len(only_marks))
```

```python
some_str = 'maths 98 english 77 science 60'
marks = []
for word in some_str.split():
    if word.isdigit():
        marks.append(int(word))
total_score = sum(marks)
average_marks = total_score / len(marks)
print("Total score:", total_score)
print("Average marks:", average_marks)
```
