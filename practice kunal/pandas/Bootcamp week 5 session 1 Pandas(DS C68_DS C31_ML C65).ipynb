```python
# merge 
```

```python
stu_details = {'id':[1,2,3,4,5],'name':['P','Q','R','S','T'],
               'region':['N','S','W','W','S']}
stu_marks = {'id':[1,2,3,4,6],'sub':['Py','matlab','R','Py','Py'],
               'marks':[10,13,14,16,13]}
```

```python
df1 = pd.DataFrame(stu_details)
df2 = pd.DataFrame(stu_marks)
```

```python
df1
```

```python
df2
```

```python
# merge
# by default inner merge
pd.merge(df1,df2) 
```

```python

pd.merge(df1,df2,on = 'id') 
```

```python
# inner merge : only common elements from both the tables
```

```python
# left : fetch everything from left table but common one from right table
pd.merge(df1,df2,on = 'id',how = 'left') 
```

```python
# right : fetch everything from right table but common one from left table
pd.merge(df1,df2,on = 'id',how = 'right') 
```

```python
# outer : union of both the tables / df
pd.merge(df1,df2,on = 'id',how = 'outer') 
```

```python
# what if common column name are diff in both the tables
```
